```python
import numpy as np
import pandas as pd


df= pd.read_csv('/content/drive/MyDrive/full_moviess.csv')

df
```

```python
df.info()
```

```python
df.dropna(subset=['overview'], inplace=True)
display(df.info())
```

```python
def make_embedding_text(row):
    parts = []

    if pd.notna(row['title']):
        parts.append(f"Movie Title: {row['title']}")

    if pd.notna(row['overview']):
        parts.append(f"Movie Overview/Plot: {row['overview']}")

    if pd.notna(row['cast']):
        parts.append(f"Cast: {row['cast']}")

    if pd.notna(row['genres']):
        parts.append(f"Genres: {row['genres']}")

    if pd.notna(row['director']):
        parts.append(f"Director: {row['director']}")

    return ". ".join(parts)
```

```python
df["text_for_embedding"]=df.apply(make_embedding_text,axis=1)
```

```python
from sentence_transformers import SentenceTransformer

# Load embedding model
model = SentenceTransformer("all-mpnet-base-v2")

# Generate embeddings
embeddings = model.encode(df["text_for_embedding"].tolist(), show_progress_bar=True)
```

```python
embeddings
```

```python
import chromadb


# Connect to Chroma
# client = chromadb.PersistentClient(path="/content/drive/MyDrive/chroma_db")
client = chromadb.PersistentClient(path="/content/chroma_db1")
collection = client.create_collection("movies_embeddings", get_or_create=True)

# Insert into Chroma
for i, row in df.iterrows():
    collection.add(
        ids=[str(i)],
        embeddings=[embeddings[i]],
        metadatas=[{
            "movie_id": row["id"],
            "title": row["title"],
            "release_date": row["release_date"],
            "genres": row["genres"],
            "director": row["director"],
            "cast": row["cast"],
            "poster_path": row["poster_path"]
        }],
        documents=[row["text_for_embedding"]]
    )
```

```python
import chromadb


# Connect to Chroma (in-memory)
client = chromadb.Client()  # In-memory client
collection = client.create_collection("movies_embeddings", get_or_create=True)

# Prepare batch data
ids = [str(i) for i in range(len(df))]
embeddings_list = embeddings.tolist()  # Convert numpy array to list
metadatas = [
    {
        "movie_id": row["id"],
        "title": row["title"],
        "release_date": row["release_date"],
        "genres": row["genres"],
        "director": row["director"],
        "cast": row["cast"],
        "poster_path": row["poster_path"]
    } for _, row in df.iterrows()
]
documents = df["text_for_embedding"].tolist()

# Define batch size (must be <= 5461)
batch_size = 5000  # Choose a size below the max limit (5461)

# Insert data in batches
for start_idx in range(0, len(df), batch_size):
    end_idx = min(start_idx + batch_size, len(df))
    batch_ids = ids[start_idx:end_idx]
    batch_embeddings = embeddings_list[start_idx:end_idx]
    batch_metadatas = metadatas[start_idx:end_idx]
    batch_documents = documents[start_idx:end_idx]

    # Insert batch into Chroma
    collection.add(
        ids=batch_ids,
        embeddings=batch_embeddings,
        metadatas=batch_metadatas,
        documents=batch_documents
    )
    print(f"Inserted batch {start_idx} to {end_idx}")
```

```python
from sentence_transformers import SentenceTransformer
import chromadb

# Load the embedding model (same as used for creating embeddings)
model = SentenceTransformer("all-mpnet-base-v2")

# Connect to Chroma (in-memory or persistent)
client = chromadb.Client()  # Use PersistentClient if you saved to disk
collection = client.get_collection("movies_embeddings")

# Define the query
query_text = input("Enter your query")


# Generate embedding for the query
query_embedding = model.encode([query_text])[0].tolist()

# Perform similarity search
results = collection.query(
    query_embeddings=[query_embedding],
    n_results=5  # Return top 2 results (equivalent to k=2)
)

# Process and display results
for i, (id, document, metadata, distance) in enumerate(zip(
    results["ids"][0], results["documents"][0], results["metadatas"][0], results["distances"][0]
)):
    print(f"Result {i+1}:")
    print(f"ID: {id}")
    print(f"Title: {metadata['title']}")
    print(f"Document: {document}")
    print(f"Distance: {distance}")
    print(f"Metadata: {metadata}")
    print("---")
```
