```python
import pandas as pd
import requests


all_movies = []

# TMDB allows max 500 pages for popular movies
for j in range(1, 501):
    url = f"https://api.themoviedb.org/3/discover/movie?api_key={api_key}&language=en-US&with_origin_country=IN&page={j}"
    response = requests.get(url).json()

    if "results" in response:
        all_movies.extend(response["results"])
    else:
        print(f"Error on page {j}: {response}")

# convert to DataFrame
final = pd.DataFrame(all_movies)
print(final.head())
```

```python
url1 = f"https://api.themoviedb.org/3/genre/movie/list?api_key={api_key}&language=en-US"
genres_response = requests.get(url1).json()

# create dict {id: name}

genre_dict = {g["id"]: g["name"] for g in genres_response["genres"]}
```

```python
genre_dict
```

```python
def decode_genres(genre_ids):
    return [genre_dict.get(gid, "Unknown") for gid in genre_ids]

final["genres"] = final["genre_ids"].apply(decode_genres)
```

```python
def get_cast_director(movie_id):
    url = f"https://api.themoviedb.org/3/movie/{movie_id}/credits?api_key={api_key}&language=en-US"
    credits = requests.get(url).json()
    cast = [c["name"] for c in credits.get("cast", [])[:3]]
    director = [c["name"] for c in credits.get("crew", []) if c["job"] == "Director"]
    return {"cast": cast, "director": director}

# Example: add columns to DataFrame
final["credits"] = final["id"].apply(get_cast_director)
final["cast"] = final["credits"].apply(lambda x: x["cast"])
final["director"] = final["credits"].apply(lambda x: x["director"])

```
